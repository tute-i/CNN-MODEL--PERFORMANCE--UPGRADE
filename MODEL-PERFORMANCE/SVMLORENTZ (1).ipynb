```python
import torch
from torchvision import datasets, transforms
from sklearn import svm, metrics

# Veri dönüşümleri
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.1307,), (0.3081,))
])

# MNIST veri setini yükleme
train_dataset = datasets.MNIST(root='./data', train=True, download=True, transform=transform)
test_dataset = datasets.MNIST(root='./data', train=False, download=True, transform=transform)

# Veri yükleyicilerini oluşturma
batch_size = 64
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

# SVM modelini oluşturma
model = svm.SVC()

# Her bir iterasyonda eğitim verilerini kullanarak SVM modelini eğitme
num_epochs = 2
for epoch in range(num_epochs):
    train_data = []
    train_labels = []
    for data, labels in train_loader:
        data = data.reshape(-1, 28*28)
        train_data.append(data)
        train_labels.append(labels)
    train_data = torch.cat(train_data, dim=0)
    train_labels = torch.cat(train_labels, dim=0)
    model.fit(train_data, train_labels)

    # Test verilerini tahmin etme
    test_data = []
    test_labels = []
    for data, labels in test_loader:
        data = data.reshape(-1, 28*28)
        test_data.append(data)
        test_labels.append(labels)
    test_data = torch.cat(test_data, dim=0)
    test_labels = torch.cat(test_labels, dim=0)
    test_predictions = model.predict(test_data)

    # Sınıflandırma raporunu yazdırma
    classification_report = metrics.classification_report(test_labels, test_predictions)
    print(f"Iterasyon {epoch+1}/{num_epochs} - Sınıflandırma Raporu:")
    print(classification_report)
```

```python
def apply_lorentz_transformation(data):
    # Lorentz dönüşümü uygulama kodu
    beta = 0.99
    gamma = 1 / (torch.sqrt(torch.tensor(1 - beta**2)) if beta < 1 else float('inf'))  # Lorentz faktörü

    data_tensor = torch.tensor(data, dtype=torch.float32)  # Data'yı Tensor'a dönüştür
    transformed_data = gamma * (data_tensor - beta * data_tensor)
    return transformed_data
```

```python
import torch
from torchvision import datasets, transforms
from sklearn import svm, metrics

# Veri dönüşümleri
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.1307,), (0.3081,))
])

# MNIST veri setini yükleme
train_dataset = datasets.MNIST(root='./data', train=True, download=True, transform=transform)
test_dataset = datasets.MNIST(root='./data', train=False, download=True, transform=transform)

# Veri yükleyicilerini oluşturma
batch_size = 64
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

# SVM modelini oluşturma
model = svm.SVC()

# Her bir iterasyonda eğitim verilerini kullanarak SVM modelini eğitme
num_epochs = 2
for epoch in range(num_epochs):
    train_data = []
    train_labels = []
    for data, labels in train_loader:
        data = data.reshape(-1, 28*28)
        transformed_data = apply_lorentz_transformation(data)  # Lorentz dönüşümü uygula
        train_data.append(transformed_data)
        train_labels.append(labels)
    train_data = torch.cat(train_data, dim=0)
    train_labels = torch.cat(train_labels, dim=0)
    model.fit(train_data, train_labels)

    # Test verilerini tahmin etme
    test_data = []
    test_labels = []
    for data, labels in test_loader:
        data = data.reshape(-1, 28*28)
        transformed_data = apply_lorentz_transformation(data)  # Lorentz dönüşümü uygula
        test_data.append(transformed_data)
        test_labels.append(labels)
    test_data = torch.cat(test_data, dim=0)
    test_labels = torch.cat(test_labels, dim=0)
    test_predictions = model.predict(test_data)

    # Sınıflandırma raporunu yazdırma
    classification_report = metrics.classification_report(test_labels, test_predictions)
    print(f"Iterasyon {epoch+1}/{num_epochs} - Sınıflandırma Raporu:")
    print(classification_report)
```
